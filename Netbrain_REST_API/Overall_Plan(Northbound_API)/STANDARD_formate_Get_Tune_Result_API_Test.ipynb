# Tune Device API Design

## ***GET*** /V1/CMDB/TuneDevices{?TaskId}&{?Begin}&{?Count}&{?Hostnames}
Calling this API to start a tune devies task.

* ## Detail Information

> **Title** : Get Tune Result API<br>

> **Version** : 02/06/2019.

> **API Server URL** : http(s):// IP address of your NetBrain Web API Server /ServicesAPI/API/V1/CMDB/TuneDevices

> **Authentication** : 

>|**Type**|**In**|**Name**|
|------|------|------|
|<img width=100/>|<img width=100/>|<img width=500/>|
|Bearer Authentication| Parameters | Authentication token | 

* ## Request body(****required***)

>No request body.

* ## Query Parameters(****required***)

> |**Name**|**Type**|**Description**|
|------|------|------|
|<img width=100/>|<img width=100/>|<img width=500/>|
| TaskId | string  | The tune task id. Users need to get the TaskId by calling "Tune Devices API"|
| Begin | int | Begin index of data, API will return device tune result begin at "Begin". |
| Count | int  | Number of returned data, API will return device tune result, the total number equal to "count". |
| Hostnames | list of string | List all devices that need to return tune result.<br>If Hostnames has value then Begin and Count are useless;<br>If Hostnmaes has no value then, we use Begin and Count to return tune results. |
